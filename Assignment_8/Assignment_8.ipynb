```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
print("types of Distribution plot are below :")
print("1) Dist-plot")
sns.distplot(x=dataset['age'], bins=10)
```

```python
sns.distplot(x=dataset['age'],bins=10, kde=False)
```

```python
print("2) joint-plot")
sns.jointplot(x=dataset['age'], y=dataset['fare'], data=dataset, kind='scatter')
```

```python
sns.jointplot(x=dataset['age'], y=dataset['fare'], data=dataset, kind='hex')
```

```python
print("3) rug plot")
sns.rugplot(dataset['fare'])
```

```python
print("types of Categorical Plots are below :")
print("1) bar-plot")
sns.barplot(x=dataset['sex'], y=dataset['age'], data=dataset)
```

```python
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
```

```python
print("2) count-plot")
sns.countplot(x=dataset['sex'], data=dataset)
```

```python
print("3) box-plot")
sns.boxplot(x=dataset['sex'], y=dataset['age'], data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
print("4) violin-plot")
sns.violinplot(x=dataset['sex'], y=dataset['age'], data=dataset)
```

```python
sns.violinplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
print("types of Advanced Plots are below :")
print("1) strip-plot")
sns.stripplot(x=dataset['sex'], y=dataset['age'], data=dataset, jitter=False)
```

```python
sns.stripplot(x=dataset['sex'], y=dataset['age'], data=dataset, jitter=True)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived')
```

```python
print("2) swarm-plot")
sns.swarmplot(x=dataset['sex'], y=dataset['age'], data=dataset)
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
print("types of matrix plots are below :")
print("1) heat-map")
numeric_dataset = dataset.select_dtypes(include=np.number)
correlation_matrix = numeric_dataset.corr()
sns.heatmap(correlation_matrix, annot=False)
```

```python
sns.heatmap(correlation_matrix, annot=True)
```

```python
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
```

```python
print("2) cluster-map")
sns.clustermap(correlation_matrix)
```

```python
print("Q2) checking how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.")
sns.distplot(x=dataset['fare'], bins=10, kde=False)
```
