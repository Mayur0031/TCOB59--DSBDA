```python
import pandas as pd
import numpy as np

df = pd.read_csv("/HR_Dataset.csv")
df.head()
```

```python
print(f"mean of monthly income of employees is : {df['MonthlyIncome'].mean()}")
print(f"median of monthly income of employees is : {df['MonthlyIncome'].median()}")
print(f"mode of monthly income of employees is : {df['MonthlyIncome'].mode()}")
```

```python
print(f"mean of Ages of all employees is {df['Age'].mean()}")
print(f"median of Ages of all employees is {df['Age'].median()}")
print(f"mode of Ages of all employees is {df['Age'].mode()}")
```

```python
print(f"standard deviation of monthly income of all employees is {df['MonthlyIncome'].std()}")
print(f"variance of monthly income of all employees is {df['MonthlyIncome'].var()}")
```

```python
print(f"standard deviation of Age of all employees is {df['Age'].std()}")
print(f"variance of Age of all employees is {df['Age'].var()}")
```
